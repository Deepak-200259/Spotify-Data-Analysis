```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks=pd.read_csv('D:/Project/Spotify Data Analysis Project in Python/Datasets/Spotify/tracks.csv')
df_tracks.head()
```

```python
#Check for null values
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
sorted_df=df_tracks.sort_values('popularity',ascending=True).head(10)
sorted_df
```

```python
df_tracks.describe().transpose()
```

```python
most_popular=df_tracks.query('popularity>90',inplace=False).sort_values('popularity', ascending=False).head(10)
most_popular[:10]
```

```python
df_tracks.set_index("release_date",inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head(5)
```

```python
df_tracks[['artists']].iloc[18]
```

```python
df_tracks['duration']=df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms",inplace=True,axis=1)
df_tracks.duration.head()
```

```python
#correlation map
corr_df=df_tracks.drop(['key','mode','explicit'],axis=1).corr(method="pearson")
plt.figure(figsize=(14,7))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap="inferno",linewidths=1,linecolor="Black")
heatmap.set_title("Correlation HeatMap between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)


#Conclusion
#There is strong Positive correlation i.e.0.8 between Loudness and Energy which means that the more the energy the louder will be the song or vice versa
#There is strong Negative correlation ie.-0.7 between Energy and acousticness
#Moderate Positive correlation is there i.e. 0.5 between Danceability and valence
```

```python
sample_df=df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y='loudness', x='energy', color='c').set(title="Loudness vs Energy Correlation")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y='popularity', x='acousticness', color='b').set(title="Popularity vs Acousticness Correlation")
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
#pip install --user seaborn==0.11.0
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind='hist').set(title='Number of Songs per Year')
```

```python
#Conclusion- 
#In 1920's the songs were not so popular and also a few songs were released
#From 1950 to 2000 the songs started growing and releasing at a rapid rate from 2000 songs per year to approx. 12500 per year
#After 2000, there was a downfall of songs per year which gradually recovered and still recovering.
#The Maximum songs were released in 2020 i.e. more than 14000 songs per year
```

```python
total_dr=df_tracks.duration
fig_dims=(18,7)
fig, ax=plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years,y=total_dr,ax=ax,errwidth=False).set(title='Years vs Duration')
plt.xticks(rotation=90)
```

```python
total_dr=df_tracks.duration
sns.set_style(style='whitegrid')
fig_dims=(10,5)
fig, ax=plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years,y=total_dr,ax=ax).set(title='Years vs Duration')
plt.xticks(rotation=90)
```

```python
#Conclusion-
#from 1900 to 1923, the duration of songs decreased rapidly from approx 240 ms to approx 160 ms.
#In year 1935, the duration of songs was maximum i.e. 260ms.
#Till Late 1940's, the duration of songs was less than the song released thereafter from 1940's to 2014.
#Till 2014, there were ups and downs in the duration of songs and no rapid increase or decrease in duration occured.
#After 2014, the duration of songs started decreasing from 240ms to 200 ms.
```

```python
df_genre=pd.read_csv('D:/Project/Spotify Data Analysis Project in Python/Datasets/Spotify Features/SpotifyFeatures.csv')
```

```python
plt.title('Duration of the songs in Different Genres')
sns.color_palette('rocket',as_cmap=True)
fig=sns.barplot(x='genre',y='duration_ms',data=df_genre)
plt.xlabel('Genres')
plt.ylabel('Duration in milli seconds')
plt.xticks(rotation=90)

#Conclusion
#The Duration of songs of Pop Genres and World genres is more than 300000 ms that is the highest among the genres
#The Duration of songs of Children Music Genres is lowest with approx 140000 ms.
#Most types of Genres Lies between 200000 ms to 250000 ms Duration.
```
